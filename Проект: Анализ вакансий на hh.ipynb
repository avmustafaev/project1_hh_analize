# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö

1. –ù–∞—á–Ω–µ–º —Å –ø—Ä–æ—Å—Ç–æ–≥–æ - —Å –ø—Ä–∏–∑–Ω–∞–∫–∞ **"–û–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∏ –í–£–ó"**. –ï–≥–æ —Ç–µ–∫—É—â–∏–π —Ñ–æ—Ä–º–∞—Ç —ç—Ç–æ: **<–£—Ä–æ–≤–µ–Ω—å –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –≥–æ–¥ –≤—ã–ø—É—Å–∫–∞ –í–£–ó —Å–ø–µ—Ü–∏–∞–ª—å–Ω–æ—Å—Ç—å...>**. –ù–∞–ø—Ä–∏–º–µ—Ä:
* –í—ã—Å—à–µ–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ 2016 –ú–æ—Å–∫–æ–≤—Å–∫–∏–π –∞–≤–∏–∞—Ü–∏–æ–Ω–Ω—ã–π –∏–Ω—Å—Ç–∏—Ç—É—Ç (–Ω–∞—Ü–∏–æ–Ω–∞–ª—å–Ω—ã–π –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π —É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç)...
* –ù–µ–æ–∫–æ–Ω—á–µ–Ω–Ω–æ–µ –≤—ã—Å—à–µ–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ 2000  –ë–∞–ª—Ç–∏–π—Å–∫–∞—è –≥–æ—Å—É–¥–∞—Ä—Å—Ç–≤–µ–Ω–Ω–∞—è –∞–∫–∞–¥–µ–º–∏—è —Ä—ã–±–æ–ø—Ä–æ–º—ã—Å–ª–æ–≤–æ–≥–æ —Ñ–ª–æ—Ç–∞‚Ä¶
–ù–∞—Å –±—É–¥–µ—Ç –∏–Ω—Ç–µ—Ä–µ—Å–æ–≤–∞—Ç—å —Ç–æ–ª—å–∫–æ —É—Ä–æ–≤–µ–Ω—å –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è.

–°–æ–∑–¥–∞–π—Ç–µ —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏-–ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ **"–û–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ"**, –∫–æ—Ç–æ—Ä—ã–π –¥–æ–ª–∂–µ–Ω –∏–º–µ—Ç—å 4 –∫–∞—Ç–µ–≥–æ—Ä–∏–∏: "–≤—ã—Å—à–µ–µ", "–Ω–µ–æ–∫–æ–Ω—á–µ–Ω–Ω–æ–µ –≤—ã—Å—à–µ–µ", "—Å—Ä–µ–¥–Ω–µ–µ —Å–ø–µ—Ü–∏–∞–ª—å–Ω–æ–µ" –∏ "—Å—Ä–µ–¥–Ω–µ–µ".

–í—ã–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ, –æ—Ç–≤–µ—Ç—å—Ç–µ –Ω–∞ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ã–µ –≤–æ–ø—Ä–æ—Å—ã –∏ —É–¥–∞–ª–∏—Ç–µ –ø—Ä–∏–∑–Ω–∞–∫ "–û–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∏ –í–£–ó".

–°–æ–≤–µ—Ç: –æ–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—É —Ç–µ–∫—Å—Ç–∞ –≤ —Å—Ç–æ–ª–±—Ü–µ **"–û–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∏ –í–£–ó"**. –ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —Ç–µ–∫—É—â–∏–π —É—Ä–æ–≤–µ–Ω—å –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è —Å–æ–∏—Å–∫–∞—Ç–µ–ª—è –≤—Å–µ–≥–¥–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ –ø–µ—Ä–≤—ã—Ö 2—É—Ö —Å–ª–æ–≤ –∏ –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å –∑–∞–≥–ª–∞–≤–Ω–æ–π –±—É–∫–≤—ã. –í–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å —ç—Ç–∏–º.

*–°–æ–≤–µ—Ç: –ø—Ä–æ–≤–µ—Ä—è–π—Ç–µ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏, –Ω–∞–ø—Ä–∏–º–µ—Ä, —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ unique()*


```python
# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ —Å–ø–∏—Å–æ–∫, –µ—Å–ª–∏ –≤—Ç–æ—Ä–æ–µ —Å–ª–æ–≤–æ '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ', –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –æ–¥–Ω–æ —Å–ª–æ–≤–∞, –µ—Å–ª–∏ –Ω–µ—Ç, —Å—Ä–µ–∑ –¥–≤—É—Ö –ø–µ—Ä–≤—ã—Ö —Å–ª–æ–≤.
hh_data['–û–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ'] = hh_data['–û–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∏ –í–£–ó'].apply(lambda x: (x.split()[0]).lower() if x.split()[1] == '–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ' else (" ".join(x.split()[:2])).lower())
hh_data = hh_data.drop(columns='–û–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∏ –í–£–ó', axis=1)

# —Å—á–∏—Ç–∞–µ–º value_counts, –≤–æ–∑—Ä–∞—â–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ —Å–µ—Ä–∏–∏ –ø–æ –∫–ª—é—á—É: '—Å—Ä–µ–¥–Ω–µ–µ'
school = hh_data['–û–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ'].value_counts()['—Å—Ä–µ–¥–Ω–µ–µ']
print(f'üåü {school} —Å–æ–∏—Å–∫–∞—Ç–µ–ª–µ–π –∏–º–µ—é—Ç —Ç–æ–ª—å–∫–æ —à–∫–æ–ª—å–Ω–æ–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ.')
```

2. –¢–µ–ø–µ—Ä—å –Ω–∞—Å –∏–Ω—Ç–µ—Ä–µ—Å—É–µ—Ç —Å—Ç–æ–ª–±–µ—Ü **"–ü–æ–ª, –≤–æ–∑—Ä–∞—Å—Ç"**. –°–µ–π—á–∞—Å –æ–Ω –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω –≤ —Ñ–æ—Ä–º–∞—Ç–µ **<–ü–æ–ª , –≤–æ–∑—Ä–∞—Å—Ç , –¥–∞—Ç–∞ —Ä–æ–∂–¥–µ–Ω–∏—è >**. –ù–∞–ø—Ä–∏–º–µ—Ä:
* –ú—É–∂—á–∏–Ω–∞ , 39 –ª–µ—Ç , —Ä–æ–¥–∏–ª—Å—è 27 –Ω–æ—è–±—Ä—è 1979 
* –ñ–µ–Ω—â–∏–Ω–∞ , 21 –≥–æ–¥ , —Ä–æ–¥–∏–ª–∞—Å—å 13 —è–Ω–≤–∞—Ä—è 2000
–ö–∞–∫ –≤—ã –ø–æ–Ω–∏–º–∞–µ—Ç–µ, –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–¥–µ–ª–∏—Ç—å –∫–∞–∂–¥—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü.

–°–æ–∑–¥–∞–π—Ç–µ –¥–≤–∞ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞ **"–ü–æ–ª"** –∏ **"–í–æ–∑—Ä–∞—Å—Ç"**. –ü—Ä–∏ —ç—Ç–æ–º –≤–∞–∂–Ω–æ —É—á–µ—Å—Ç—å:
* –ü—Ä–∏–∑–Ω–∞–∫ –ø–æ–ª–∞ –¥–æ–ª–∂–µ–Ω –∏–º–µ—Ç—å 2 —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å—Ç—Ä–æ–∫–æ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è: '–ú' - –º—É–∂—á–∏–Ω–∞, '–ñ' - –∂–µ–Ω—â–∏–Ω–∞. 
* –ü—Ä–∏–∑–Ω–∞–∫ –≤–æ–∑—Ä–∞—Å—Ç–∞ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω —Ü–µ–ª—ã–º–∏ —á–∏—Å–ª–∞–º–∏.

–í—ã–ø–æ–ª–Ω–∏—Ç–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ, –æ—Ç–≤–µ—Ç—å—Ç–µ –Ω–∞ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω—ã–µ –≤–æ–ø—Ä–æ—Å—ã –∏ —É–¥–∞–ª–∏—Ç–µ –ø—Ä–∏–∑–Ω–∞–∫ **"–ü–æ–ª, –≤–æ–∑—Ä–∞—Å—Ç"** –∏–∑ —Ç–∞–±–ª–∏—Ü—ã.

*–°–æ–≤–µ—Ç: –æ–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—É —Ç–µ–∫—Å—Ç–∞ –≤ —Å—Ç–æ–ª–±—Ü–µ, –≤ —á–∞—Å—Ç–∏ –Ω–∞ —Ç–æ, –∫–∞–∫ —Ä–∞–∑–¥–µ–ª–µ–Ω—ã –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –ø–æ–ª–∞, –≤–æ–∑—Ä–∞—Å—Ç–∞ –∏ –¥–∞—Ç—ã —Ä–æ–∂–¥–µ–Ω–∏—è –º–µ–∂–¥—É —Å–æ–±–æ–π - —Å–∏–º–≤–æ–ª–æ–º ' , '. 
–ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ –æ–¥–∏–Ω–∞–∫–æ–≤–∞ –¥–ª—è –≤—Å–µ—Ö —Å—Ç—Ä–æ–∫ –≤ —Ç–∞–±–ª–∏—Ü–µ. –í—ã –º–æ–∂–µ—Ç–µ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —ç—Ç–∏–º.*


```python
# –ø—Ä–æ—Å—Ç–æ –±–µ—Ä—ë–º –ø–µ—Ä–≤—É—é –±—É–∫–≤—É.
hh_data['–ü–æ–ª'] = hh_data['–ü–æ–ª, –≤–æ–∑—Ä–∞—Å—Ç'].apply(lambda x: x[0])

# —Å–ø–ª–∏—Ç–∏–º –ø–æ –º–∞—Å–∫–µ: –ø—Ä–æ–±–µ–ª –∑–∞–ø—è—Ç–∞—è –ø—Ä–æ–±–µ–ª, –¥–æ—Å—Ç–∞—ë–º –≤—Ç–æ—Ä–æ–µ –≤—Ö–æ–∂–¥–µ–Ω–∏–µ –≤ —Å–ø–∏—Å–æ–∫, —Å–ø–ª–∏—Ç–∏–º —Å–Ω–æ–≤–∞ –¥–æ—Å—Ç–∞—ë–º –ø–µ—Ä–≤–æ–µ –≤—Ö–æ–∂–¥–µ–Ω–∏–µ.
hh_data['–í–æ–∑—Ä–∞—Å—Ç'] = hh_data['–ü–æ–ª, –≤–æ–∑—Ä–∞—Å—Ç'].apply(lambda x: int(x.split(' , ')[1].split()[0]))

hh_data = hh_data.drop('–ü–æ–ª, –≤–æ–∑—Ä–∞—Å—Ç', axis=1)

# —Å—á–∏—Ç–∞–µ–º value_counts —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º normalize, —É–º–Ω–æ–∂–∞–µ–º –∏ –æ–∫—Ä—É–≥–ª—è–µ–º, –≤–æ–∑–≤—Ä–∞—â–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ —Å–µ—Ä–∏–∏ –ø–æ –∫–ª—é—á—É '–ñ'
w_persent = (hh_data['–ü–æ–ª'].value_counts(normalize=True) * 100).round(2)['–ñ']
print(f'üåü {w_persent}% –∂–µ–Ω—Å–∫–∏—Ö —Ä–µ–∑—é–º–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–æ –≤ –¥–∞–Ω–Ω—ã—Ö.')

# –±–µ—Ä—ë–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –æ–∫—Ä—É–≥–ª—è–µ–º
mean_age = hh_data['–í–æ–∑—Ä–∞—Å—Ç'].mean().round(1)
print(f'üåü –°—Ä–µ–¥–Ω–∏–π –≤–æ–∑—Ä–∞—Å—Ç —Å–æ–∏—Å–∫–∞—Ç–µ–ª–µ–π: {mean_age}.')
```

3. –°–ª–µ–¥—É—é—â–∏–º —ç—Ç–∞–ø–æ–º –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –ø—Ä–∏–∑–Ω–∞–∫ **"–û–ø—ã—Ç —Ä–∞–±–æ—Ç—ã"**. –ï–≥–æ —Ç–µ–∫—É—â–∏–π —Ñ–æ—Ä–º–∞—Ç - —ç—Ç–æ: **<–û–ø—ã—Ç —Ä–∞–±–æ—Ç—ã: n –ª–µ—Ç m –º–µ—Å—è—Ü–µ–≤, –ø–µ—Ä–∏–æ–¥—ã —Ä–∞–±–æ—Ç—ã –≤ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∫–æ–º–ø–∞–Ω–∏—è—Ö‚Ä¶>**. 

–ò–∑ —Å—Ç–æ–ª–±—Ü–∞ –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–¥–µ–ª–∏—Ç—å –æ–±—â–∏–π –æ–ø—ã—Ç —Ä–∞–±–æ—Ç—ã —Å–æ–∏—Å–∫–∞—Ç–µ–ª—è –≤ –º–µ—Å—è—Ü–∞—Ö, –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ –Ω–∞–∑–æ–≤–µ–º "–û–ø—ã—Ç —Ä–∞–±–æ—Ç—ã (–º–µ—Å—è—Ü)"

–î–ª—è –Ω–∞—á–∞–ª–∞ –æ–±—Å—É–¥–∏–º —É—Å–ª–æ–≤–∏—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏:

```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ —Ñ–ª–∞–≥–∞ –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç–∏ –∫ –ø–µ—Ä–µ–µ–∑–¥—É
def get_ready2move(item):
    # –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–æ–π –ø—Ä–æ–≤–µ—Ä—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏
    if '–≥–æ—Ç–æ–≤ –∫ –ø–µ—Ä–µ–µ–∑–¥—É' not in item and '–≥–æ—Ç–æ–≤–∞ –∫ –ø–µ—Ä–µ–µ–∑–¥—É' not in item and '—Ö–æ—á—É –ø–µ—Ä–µ–µ—Ö–∞—Ç—å' not in item:
        return False
    # –∑–¥–µ—Å—å –ø—Ä–æ–≤–µ—Ä—è–µ–º –≤—Ö–æ–∂–¥–µ–Ω–∏–µ –æ—Ç—Ä–∏—Ü–∞–Ω–∏—è –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç–∏
    return '–Ω–µ –≥–æ—Ç–æ–≤ –∫ –ø–µ—Ä–µ–µ–∑–¥—É' not in item and '–Ω–µ –≥–æ—Ç–æ–≤–∞ –∫ –ø–µ—Ä–µ–µ–∑–¥—É' not in item


# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ —Ñ–ª–∞–Ω–≥–∞ –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç–∏ –∫ –∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∞–º
def get_ready4trips(item):
    # –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–æ–π –ø—Ä–æ–≤–µ—Ä—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏
    if '–∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∞–º' not in item:
        return False
    # –∑–¥–µ—Å—å –ø—Ä–æ–≤–µ—Ä—è–µ–º –≤—Ö–æ–∂–¥–µ–Ω–∏–µ –æ—Ç—Ä–∏—Ü–∞–Ω–∏—è –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç–∏
    return '–Ω–µ –≥–æ—Ç–æ–≤ –∫ –∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∞–º' not in item and '–Ω–µ –≥–æ—Ç–æ–≤–∞ –∫ –∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∞–º' not in item
    
    
# —Å–æ–∑–¥–∞—ë–º —Å–ª–æ–≤–∞—Ä—å –¥–≤—É—Ö —Å—Ç–æ–ª–∏—Ü    
cap_city = {
    '–ú–æ—Å–∫–≤–∞' : '–ú–æ—Å–∫–≤–∞',
    '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥' : '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥'
    }
# —Å–ø–∏—Å–æ–∫ –º–∏–ª–ª–∏–æ–Ω–Ω–∏–∫–æ–≤
mln_cities = [
    '–ù–æ–≤–æ—Å–∏–±–∏—Ä—Å–∫', 
    '–ï–∫–∞—Ç–µ—Ä–∏–Ω–±—É—Ä–≥',
    '–ù–∏–∂–Ω–∏–π –ù–æ–≤–≥–æ—Ä–æ–¥',
    '–ö–∞–∑–∞–Ω—å', 
    '–ß–µ–ª—è–±–∏–Ω—Å–∫',
    '–û–º—Å–∫', 
    '–°–∞–º–∞—Ä–∞', 
    '–†–æ—Å—Ç–æ–≤-–Ω–∞-–î–æ–Ω—É', 
    '–£—Ñ–∞', 
    '–ö—Ä–∞—Å–Ω–æ—è—Ä—Å–∫', 
    '–ü–µ—Ä–º—å', 
    '–í–æ—Ä–æ–Ω–µ–∂',
    '–í–æ–ª–≥–æ–≥—Ä–∞–¥'
    ]
# –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å—Ç—Ä–æ–∫—É –≤ —Å–ø–∏—Å–æ–∫ –±–µ—Ä—ë–º –ø–µ—Ä–≤–æ–µ –≤—Ö–æ–∂–¥–µ–Ω–∏–µ
# –µ—Å–ª–∏ –≤—Ö–æ–¥–∏—Ç –≤ –≥–æ—Ä–æ–¥–∞ –º–∏–ª–ª–∏–æ–Ω–Ω–∏–∫–∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º '–≥–æ—Ä–æ–¥-–º–∏–ª–ª–∏–æ–Ω–Ω–∏–∫' –¥–∞–ª–µ–µ –ø—Ä–æ–≤–µ—Ä—è–µ–º —Å–ª–æ–≤–∞—Ä—å —Å—Ç–æ–ª–∏—Ü –µ—Å–ª–∏ –Ω–µ—Ç –∫–ª—é—á–∞ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º '–¥—Ä—É–≥–∏–µ', –≤—Å—ë –ø—Ä–æ—Å—Ç–æ :)
hh_data['–ì–æ—Ä–æ–¥'] = hh_data['–ì–æ—Ä–æ–¥, –ø–µ—Ä–µ–µ–∑–¥, –∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∏'].apply(lambda x: '–≥–æ—Ä–æ–¥-–º–∏–ª–ª–∏–æ–Ω–Ω–∏–∫' if x.split(' , ')[0] in mln_cities else cap_city.get(x.split(' , ')[0], '–¥—Ä—É–≥–∏–µ'))

# —Å–æ–∑–¥–∞—ë–º –ø—Ä–∏–∑–Ω–∞–∫–∏ —Ñ–ª–∞–∂–∫–∏ –º–∏–≥–∞–ª–∫–∏ —Å –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏
hh_data['–ì–æ—Ç–æ–≤–Ω–æ—Å—Ç—å –∫ –ø–µ—Ä–µ–µ–∑–¥—É'] = hh_data['–ì–æ—Ä–æ–¥, –ø–µ—Ä–µ–µ–∑–¥, –∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∏'].apply(get_ready2move)
hh_data['–ì–æ—Ç–æ–≤–Ω–æ—Å—Ç—å –∫ –∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∞–º'] = hh_data['–ì–æ—Ä–æ–¥, –ø–µ—Ä–µ–µ–∑–¥, –∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∏'].apply(get_ready4trips)

# –¥—Ä–æ–ø–∞–µ–º –Ω–µ–Ω—É–∂–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫
hh_data = hh_data.drop('–ì–æ—Ä–æ–¥, –ø–µ—Ä–µ–µ–∑–¥, –∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∏', axis=1)

# —Å—á–∏—Ç–∞–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç –ø–æ –≥–æ—Ä–æ–¥—É –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥
spb_persent = round(hh_data['–ì–æ—Ä–æ–¥'].value_counts(normalize=True) * 100, 0)['–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥']
print(f'üåü {spb_persent}% —Å–æ–∏—Å–∫–∞—Ç–µ–ª–µ–π –∂–∏–≤—É—Ç –≤ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ')

# –∑–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Å–∫—É –∏ –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—å –∫ –ø–µ—Ä–µ–µ–µ–∑–¥—É –∏ –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç—å –∫ –∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∞–º
mask = ((hh_data['–ì–æ—Ç–æ–≤–Ω–æ—Å—Ç—å –∫ –∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∞–º'] == True) & (hh_data['–ì–æ—Ç–æ–≤–Ω–æ—Å—Ç—å –∫ –ø–µ—Ä–µ–µ–∑–¥—É'] == True))
# —Ñ–∏–ª—å—Ç—Ä—É–µ–º –ø–æ –º–∞—Å–∫–µ, –ø–æ–ª—É—á–∞–µ–º —Ä–∞–∑–º–µ—Ä, –Ω–∞—Ö–æ–¥–∏–º –ø—Ä–æ—Ü–µ–Ω—Ç –æ—Ç –æ–±—â–µ–≥–æ —Ä–∞–∑–º–µ—Ä–∞
filtered_data = round(100 * hh_data[mask].shape[0] / hh_data.shape[0], 0)
print(f'üåü {filtered_data}% —Å–æ–∏—Å–∫–∞—Ç–µ–ª–µ–π –≥–æ—Ç–æ–≤—ã –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –∏ –∫ –ø–µ—Ä–µ–µ–∑–¥–∞–º, –∏ –∫ –∫–æ–º–∞–Ω–¥–∏—Ä–æ–≤–∫–∞–º')
```

5. –†–∞—Å—Å–º–æ—Ç—Ä–∏–º –ø–æ–±–ª–∏–∂–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ **"–ó–∞–Ω—è—Ç–æ—Å—Ç—å"** –∏ **"–ì—Ä–∞—Ñ–∏–∫"**. –°–µ–π—á–∞—Å –ø—Ä–∏–∑–Ω–∞–∫–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—Ç —Å–æ–±–æ–π –Ω–∞–±–æ—Ä –∫–∞—Ç–µ–≥–æ—Ä–∏–π –∂–µ–ª–∞–µ–º–æ–π –∑–∞–Ω—è—Ç–æ—Å—Ç–∏ (–ø–æ–ª–Ω–∞—è –∑–∞–Ω—è—Ç–æ—Å—Ç—å, —á–∞—Å—Ç–∏—á–Ω–∞—è –∑–∞–Ω—è—Ç–æ—Å—Ç—å, –ø—Ä–æ–µ–∫—Ç–Ω–∞—è —Ä–∞–±–æ—Ç–∞, –≤–æ–ª–æ–Ω—Ç–µ—Ä—Å—Ç–≤–æ, —Å—Ç–∞–∂–∏—Ä–æ–≤–∫–∞) –∏ –∂–µ–ª–∞–µ–º–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞ —Ä–∞–±–æ—Ç—ã (–ø–æ–ª–Ω—ã–π –¥–µ–Ω—å, —Å–º–µ–Ω–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫, –≥–∏–±–∫–∏–π –≥—Ä–∞—Ñ–∏–∫, —É–¥–∞–ª–µ–Ω–Ω–∞—è —Ä–∞–±–æ—Ç–∞, –≤–∞—Ö—Ç–æ–≤—ã–π –º–µ—Ç–æ–¥).
–ù–∞ —Å–∞–π—Ç–µ hh.ru —Å–æ–∏—Å–∫–∞—Ç–µ–ª—å –º–æ–∂–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å —Ä–∞–∑–ª–∏—á–Ω—ã–µ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π, –Ω–∞–ø—Ä–∏–º–µ—Ä:

* —á–∞—Å—Ç–∏—á–Ω–∞—è –∑–∞–Ω—è—Ç–æ—Å—Ç—å, –ø—Ä–æ–µ–∫—Ç–Ω–∞—è —Ä–∞–±–æ—Ç–∞, –≤–æ–ª–æ–Ω—Ç–µ—Ä—Å—Ç–≤–æ

* –≤–∞—Ö—Ç–æ–≤—ã–π –º–µ—Ç–æ–¥, –≥–∏–±–∫–∏–π –≥—Ä–∞—Ñ–∏–∫, —É–¥–∞–ª–µ–Ω–Ω–∞—è —Ä–∞–±–æ—Ç–∞, –ø–æ–ª–Ω–∞—è –∑–∞–Ω—è—Ç–æ—Å—Ç—å

4. –£–º–Ω–æ–∂–∏—Ç—å —Å—É–º–º—É –∂–µ–ª–∞–µ–º–æ–π –∑–∞—Ä–∞–±–æ—Ç–Ω–æ–π –ø–ª–∞—Ç—ã –Ω–∞ –ø—Ä–∏—Å–æ–µ–¥–∏–Ω–µ–Ω–Ω—ã–π –∫—É—Ä—Å –≤–∞–ª—é—Ç—ã (close) –∏ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –Ω–∞ –ø—Ä–æ–ø–æ—Ä—Ü–∏—é (–æ–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ –ø—Ä–æ–ø—É—Å–∫–∏ –ø–æ—Å–ª–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è –≤ —ç—Ç–∏—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö), —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∑–∞–Ω–µ—Å—Ç–∏ –≤ –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü "–ó–ü (—Ä—É–±)".


```python
# —Å–æ–∑–¥–∞—ë–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∫—É—Ä—Å–∞–º–∏ –≤–∞–ª—é—Ç –ø–æ –¥–∞—Ç–∞–º
currency_data = pd.hh_data = pd.read_csv('data/ExchangeRates.csv', delimiter=',')
# –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –¥–∞—Ç—ã –≤ —Ñ–æ—Ä–º–∞—Ç datetime –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ —Å –∫—É—Ä—Å–∞–º–∏
currency_data['date'] = pd.to_datetime(currency_data['date'], dayfirst=True).dt.date
# –¥—Ä–æ–ø–∞–µ–º –ª–∏—à–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ —Å –∫—É—Ä—Å–∞–º–∏
currency_data = currency_data.drop(['per', 'vol','time'],axis=1)
# –≤ –æ—Å–Ω–æ–≤–Ω–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ —Å–æ–∑–¥–∞—ë–º –ø—Ä–∏–∑–Ω–∞–∫ date —Å—Ä–∞–∑—É –≤ —Ñ–æ—Ä–º–∞—Ç–µ datetime
hh_data['date'] = pd.to_datetime(hh_data['–û–±–Ω–æ–≤–ª–µ–Ω–∏–µ —Ä–µ–∑—é–º–µ'], dayfirst=True).dt.date
# –¥—Ä–æ–ø–∞–µ–º –Ω–µ–Ω—É–∂–Ω—ã–µ —É–∂–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
hh_data = hh_data.drop('–û–±–Ω–æ–≤–ª–µ–Ω–∏–µ —Ä–µ–∑—é–º–µ', axis=1)


# —Å—Ç—Ä–æ–∫—É —Å –∑–∞—Ä–ø–ª–∞—Ç–æ–π –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ —Å–ø–∏—Å–æ–∫ –∑–∞–±–∏—Ä–∞–µ–º –ø–µ—Ä–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ —Ü–µ–ª–æ–µ –∏ —Å–æ–∑–¥–∞—ë–º –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫
hh_data['–∑–ø —á–∏—Å–ª–æ'] = hh_data['–ó–ü'].apply(lambda x: int(x.split()[0]))

# —Å–æ–∑–¥–∞—ë–º —Å–ª–æ–≤–∞—Ä—å —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏—è –Ω–∞–∑–≤–∞–Ω–∏–π –≤–∞–ª—é—Ç –≤ –æ—Å–Ω–æ–≤–Ω–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ –∏ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ —Å –∫—É—Ä—Å–∞–º–∏
currency_dict = {
    '—Ä—É–±.': 'RUB', 
    'KZT': 'KZT', 
    'USD': 'USD', 
    '–±–µ–ª.—Ä—É–±.': 'BYN', 
    'EUR': 'EUR', 
    '–≥—Ä–Ω.': 'UAH', 
    '—Å—É–º': 'UZS', 
    'KGS': 'KGS', 
    'AZN': 'AZN'
    }
# –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –ø–æ–ª–µ –≤ —Å–ø–∏—Å–æ–∫, –±–µ—Ä—ë–º –≤—Ç–æ—Ä–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ —Å–ø–∏—Å–∫–∞, –Ω–∞ –≤—Å—è–∫–∏–π —Å–ª—É—á–∞–π —Ä—É–±–∏–º –ø—Ä–æ–±–µ–ª—ã –∏  –ø–æ –∫–ª—é—á—É –∑–∞–±–∏—Ä–∞–µ–º –∏–∑ —Å–ª–æ–≤–∞—Ä—è –Ω–∞–∑–≤–∞–Ω–∏–µ –≤–∞–ª—é—Ç—ã –≤ —Ñ–æ—Ä–º–∞—Ç–µ ISO
# —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º —Å–æ–∑–¥–∞—ë–º –ø—Ä–∏–∑–Ω–∞–∫ —Å —Ç–∞–∫–∏–º –∂–µ –Ω–∞–∑–≤–∞–Ω–∏–µ–º –∫–∞–∫ –∏ –≤–æ –≤—Ç–æ—Ä–æ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ —Å –∫—É—Ä—Å–∞–º–∏, —á—Ç–æ–±—ã –ø—Ä–æ—â–µ –æ–±—ä–µ–¥–∏–Ω—è—Ç—å
hh_data['currency'] = hh_data['–ó–ü'].apply(lambda x: currency_dict.get(x.split()[1].strip()))

# –¥—Ä–æ–ø–∞–µ–º —É–∂–µ –Ω–µ–Ω—É–∂–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫
hh_data = hh_data.drop('–ó–ü', axis=1)

# –æ–±—ä–µ–¥–∏–Ω—è–µ–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã –≥–¥–µ –æ—Å–Ω–æ–≤–Ω–∞—è —á–∞—Å—Ç—å –ª–µ–≤–∞—è —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π –ø–æ –ø—Ä–∏–∑–Ω–∞–∫–∞–º: –¥–∞—Ç–∞ –∏ –Ω–∞–∑–≤–∞–Ω–∏–µ –≤–∞–ª—é—Ç—ã
hh_merged = hh_data.merge(currency_data, how='left', on=['date','currency'])

# —Ç–∞–∫ –∫–∞–∫ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ —Å –∫—É—Ä—Å–∞–º–∏ –Ω–µ –±—ã–ª–æ —Ä—É–±–ª—è, –≤—Å–µ —Å—Ç—Ä–æ—á–∫–∏ —Å —Ä—É–±–ª—ë–º –ø—É—Å—Ç—ã–µ, –¥–µ–ª–∞–µ–º –ø—Ä–æ–≤–µ—Ä–∫—É –∏ –∑–∞–º–µ–Ω—è–µ–º NaN –Ω–∞ –µ–¥–∏–Ω–∏—Ü—É –≤ –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö close –∏ proportion
hh_merged['close'] = hh_merged['close'].apply(lambda x: 1 if pd.isna(x) else x)
hh_merged['proportion'] = hh_merged['proportion'].apply(lambda x: 1 if pd.isna(x) else x)

# —Å–æ–∑–¥–∞—ë–º –∏—Ç–æ–≥–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫ –∑–∞—Ä–ø–ª–∞—Ç–∞ –≤ —Ä—É–±–ª—è—Ö, –ø–µ—Ä–µ—Å—á–∏—Ç–∞–Ω–Ω—É—é –∏–∑ –∑–∞—Ä–ø–ª–∞—Ç—ã –≤ –≤–∞–ª—é—Ç–µ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å –∫—É—Ä—Å–æ–º –Ω–∞ –¥–µ–Ω—å –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ —Ä–µ–∑—é–º–µ
hh_merged['–ó–ü (—Ä—É–±)'] = hh_merged['–∑–ø —á–∏—Å–ª–æ'] * hh_merged['close'] / hh_merged['proportion']

# –¥—Ä–æ–ø–∞–µ–º –Ω–µ–Ω—É–∂–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ –∑–∞–º–µ–Ω—è–µ–º –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –≤ –∏–∑–Ω–∞—á–∞–ª—å–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
hh_data = hh_merged.drop(['–∑–ø —á–∏—Å–ª–æ', 'currency', 'close', 'proportion'], axis=1)

# –≤—ã—Å—á–∏—Ç—ã–≤–∞–µ–º –º–µ–¥–∏–∞–Ω–Ω—É—é –∂–µ–ª–∞–µ–º—É—é –∑–∞—Ä–ø–ª–∞—Ç—É —Å–æ–∏—Å–∫–∞—Ç–µ–ª–µ–π –≤ —Ç—ã—Å—è—á–∞—Ö —Ä—É–±–ª–µ–π
median_salary = (hh_data['–ó–ü (—Ä—É–±)'].median() / 1000).round(0)
print(f'üåü –ñ–µ–ª–∞–µ–º–∞—è –º–µ–¥–∏–∞–Ω–Ω–∞—è –∑–∞—Ä–∞–±–æ—Ç–Ω–∞—è –ø–ª–∞—Ç–∞ —Å–æ–∏—Å–∫–∞—Ç–µ–ª–µ–π –≤ –Ω–∞—à–µ–π —Ç–∞–±–ª–∏—Ü–µ —Ä–∞–≤–Ω–∞: {median_salary} —Ç—ã—Å—è—á —Ä—É–±–ª–µ–π')
hh_data
```
